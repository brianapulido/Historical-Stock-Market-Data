```python
#https://blog.quantinsti.com/stock-market-data-analysis-python/
#Import yfinance package 
import yfinance as yf
# set the start and end dates for the data you want to obtain
start_date = '2021-06-01'
end_date = '2022-04-13'
#choose the stock ticker symbol you want to obtain the data for
ticker = 'PATH'
#get the data 
data = yf.download(ticker, start_date, end_date)
#print 5 rows of data
data.tail()
```

```python
#use matplotlib to visualize the data for the date range chosen 
import matplotlib.pyplot as plt
%matplotlib inline
#plot the adjusted close price data of the stock for the date range
data['Adj Close'].plot()
plt.show()
```

```python
#improve how the plot looks with grid lines, labels and resizing
#plot the adjusted close price
data['Adj Close'].plot(figsize=(10, 7))
#define the label for the title of the figure
plt.title("Adjusted Close Price of %s" % ticker, fontsize=16)
#define the labels for the x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
#plot the grid lines 
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
#show the plot
plt.show()
```

```python
#to get the stock market data for multiple stock tickers
import pandas as pd
#create a stock ticker list
tickers_list = ['PATH', 'VZ', 'AAL', 'JBLU', 'BE']
#create a placeholder for the data using a dataframe
data = pd.DataFrame(columns=tickers_list)
#fetch the data using the yfinance download method
for ticker in tickers_list:
    data[ticker] = yf.download(ticker, start_date, end_date)['Adj Close']
#print the first 5 rows of the data
    data.head()
```

```python
#create a chart with the adjsuted close prices of the stock data for the 5 stocks picked
#plot the close prices
data.plot(figsize=(10, 7))
#show the legend
plt.legend()
#define the label for the title of the figure
plt.title("Adjusted Close Price", fontsize=16)
#define the labels for the x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
#plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
#show the plot
plt.show()
```
